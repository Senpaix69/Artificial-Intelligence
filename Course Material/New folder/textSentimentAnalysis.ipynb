```python
import numpy as np
import csv as csv
import sys
import math
import operator


sys.maxxsize = 2147483647 

#trnData = csv.reader(open('SentimentAnalysistask.csv', 'r'))
data1 = []

with open('SentimentAnalysistask.csv', 'r', encoding = 'utf8') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for pnt1 in readCSV:
        data1.append(pnt1)
        
totalpoints = len(data1)
points = [0] * totalpoints
results = [0] * totalpoints

for i in range(totalpoints):
    str1 = ''.join(data1[i])
    #print(str1)
    results[i]=str1[0:8]
    points[i] = str1[8:]

#print (results, points)    
```

```python
testPoints = points[0:400] + points[3600:]
trainPoints = points[400:3600]
testResults = results[0:400] + results[3600:]
trainResults = results[400:3600]
#print(len(testPoints),len(trainPoints),len(testResults), len(trainResults))
```

```python
trnstr = ' '.join(trainPoints)
#print(trnstr)
```

```python
from string import punctuation
import re

s = re.sub(r'[^\w\s]','',trnstr)
#print(s)
def uniqueVocabulary(str):
    return set(str.translate(punctuation).split())

uniq = uniqueVocabulary(s)
print(type(uniq), len(uniq), uniq)
```

```python
uniqVocab = len(uniq)
l = [i for i in range(uniqVocab)]
dict1 = dict.fromkeys(uniq, 0)#{}
print(len(dict1))            
```

```python
trnrslt = np.zeros(shape=(len(trainResults)))
tstrslt = np.zeros(shape=(len(testResults)))

for i in range(len(trainResults)):
    if trainResults[i] == 'Positive':
        trnrslt[i] = 1
        
for i in range(len(testResults)):
    if testResults[i] == 'Positive':
        tstrslt[i] = 1
        
#print(len(trnrslt), len(tstrslt))
```

```python

print(dict1.keys())
```

```python
i = 0
for k in dict1.items():
    temp = {}
    temp[k[0]] = i
    dict1.update(temp)
    i += 1

#print(dict1)
```

```python
#print(dict1['am'])
trnM = np.zeros(shape=(len(trainResults), uniqVocab))
i = 0
for p in trainPoints:
    s = re.sub(r'[^\w\s]','',p)
    #print (s)
    for q in s.split(' '):
        #print(q)
        if q in dict1:
            trnM[i,dict1[q]] = 1
    i += 1
```

```python
c1 = np.zeros(shape=(2,uniqVocab))
i = 0
for k in trnM:
    if(trnrslt[i] == 0):
        c1[0] = c1[0] + k
    else:
        c1[1] = c1[1] + k
        
    i += 1
    
#print(c1[1, 0:200])
```

```python
pN = 1600
pP = 1600
pt = 3200
pNs = pN+2
pPs = pP +2
probN = pN/pt
probP = pP/pt

p1 = np.zeros(shape=(4,uniqVocab))
p1[0] = (c1[0] +1) / pNs
p1[1] = (abs(c1[0] - 1600)+1)/pNs
p1[2] = (c1[1] +1) / pPs
p1[3] =(abs(c1[1] - 1600)+1)/pPs
```

```python
tstM = np.zeros(shape=(len(testResults), uniqVocab))

i = 0
for p in testPoints:
    s = re.sub(r'[^\w\s]','',p)
    #print (s)
    for q in s.split(' '):
        #print(q)
        if q in dict1:
            tstM[i,dict1[q]] = 1
    i += 1
```

```python
tsting = np.zeros(shape=(2,uniqVocab))
predicts = np.zeros(shape=(len(testResults)))
#print(len(testResults))
i = 0
for x in tstM:
    j = 0
    #print(len(tstM), len(x))
    for y in x:
        #print(y)
        if y == 1:
            tsting[0,j] = p1[0,j]
            tsting[1,j] = p1[2,j]
        else:
            tsting[0,j] = p1[1,j]
            tsting[1,j] = p1[3,j]
        j += 1
        
                
    if((np.prod(tsting[0])*pN) > (np.prod(tsting[1])*pP)):
        predicts[i] = 0
    else:
        predicts[i] = 1
    i += 1
    
#print(predicts)
```

```python
print(predicts)
```

```python
def getAccuracy(predicts, actuals):
    correct = 0
    incorrect = 0
    lng = len(predicts)
    for i in range(lng):
        if predicts[i] == actuals[i]:
            correct += 1
        
        else:
            incorrect += 1
            
    return (correct/lng)*100
```

```python
print(getAccuracy(predicts, tstrslt))
```
